# Q. What is Dogecoin?

* Dogecoin is an open source peer-to-peer digital currency, favored by Shiba Inus worldwide. 
* It is qualitatively more fun while being technically nearly identical to its close relative Bitcoin. 
* This dataset contains its historical stock price in USD on a daily frequency starting from 17 September 2014 to 05 June 2021.

* Dataset:-https://www.kaggle.com/dhruvildave/dogecoin-historical-data

#### Importing libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
doge_df=pd.read_csv('DOGE-USD.csv')
doge_df
```

### Dataset contains 2479 rows ans 7 columns

```python
doge_df.columns
```

```python
# Dropping 'Adj Close' as it represents the same data as 'Close' column
doge_df.drop(['Adj Close'],axis=1, inplace = True)
```

```python
plt.figure(figsize = (20,10))
doge_df.groupby('Date')['Open'].mean().plot(marker='o',ls='-',lw=3,ms=3,mew=1,alpha=.5)
plt.xlabel('Date')
plt.ylabel('Volume')

plt.show()
```

* Fluctuation in the prices of Dogecoin came after mid of 2020, let's explore more from mid 2020 to 2021.

```python
# So as we can see price fluctuation is after 2020, let's focus more on year 2020-2021
doge_2020_21 = doge_df.loc[doge_df['Date'] >= '2020-06-01']
doge_2020_21
```

```python
high = doge_2020_21.High
print(f'Dogecoin highest touchdown: {high.max()} \nTotal count: {doge_2020_21.High.count()}')

plt.hist(high, bins=20, histtype='bar', rwidth=.8, color='red')
plt.ylabel('Count Range')
plt.xlabel(r'Price (in \$\$)')
plt.show()
```

Last one year performance of Dogecoin: 
- majority of times it remained low 
- for around 30-40 days it remained in its golden period

Now, let's see its performance in line graph of last one year

```python
doge_2020_21.groupby("Date")['High'].mean().plot(kind="line")
plt.xticks(rotation=90)
plt.show()
```

From the above plot we can see the change in the prices of Dogecoin came after December of 2020

We now reducing our analysis from January 2021 to June 2021.

```python
doge_2021 = doge_df.loc[doge_df['Date'] >= '2021-01-01']

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)


doge_2021.groupby('Date')['High'].mean().plot(kind='line')
plt.grid(True)
plt.show()
```

Elon has started tweeting/replying about Dogecoin in december 2020 and it remained the same till april 2021.

In May 2021 he actively did some tweets on dogecoin which lead to massive rise in the price of dogecoin. Here are some:  

Because of these two tweets by Elon Musk two major spikes came in Dogecoin in the month of May

Reducing our analysis from April 2021 to 5th July 2021

```python
doge_2021_april_abv = doge_df.loc[doge_df['Date'] >= '2021-04-01']
doge_2021_april_abv
```

```python
plt.figure(figsize=(30,20))

grp_date_high = doge_2021_april_abv.groupby('Date')['High'].mean().plot(kind="line", marker='o', ls='-', lw=3, ms=12, mew=1, alpha=1)
grp_date_low = doge_2021_april_abv.groupby('Date')['Low'].mean().plot(kind="line", marker='o', ls='-', lw=3, ms=12, mew=1, alpha=1)

plt.xlabel('Date', fontsize=30)
plt.ylabel(r'USD $', fontsize=30)
plt.grid(color='g', ls='-', lw=0.8, alpha=1)

plt.tick_params(axis='x', which='major', labelsize=25, rotation= 0)
plt.tick_params(axis='y', which='major', labelsize=25)

plt.title('Dogecoin -- High -- Low -- April to July 5th 2021', fontsize= 40)

plt.legend(['High', 'Low'], fontsize=25)
plt.show()
```

### Date wise representation of above graph

```python
plt.figure(figsize=(30,20))

plt.plot(doge_2021_april_abv.Date, doge_2021_april_abv.High, marker='o', ls='-', lw=3, ms=12, mew=1, alpha=1)
plt.plot(doge_2021_april_abv.Date, doge_2021_april_abv.Low, marker='o', ls='-', lw=3, ms=12, mew=1, alpha=1)

plt.xlabel('Date', fontsize=30)
plt.ylabel('USD', fontsize=30)
plt.grid(color='g', ls='-', lw=0.5, alpha=1)

plt.tick_params(axis='x', which='major', labelsize=15, rotation=90)
plt.tick_params(axis='y', which='major', labelsize=25)

plt.title('Dogecoin -- High -- Low -- April to July 5th 2021', fontsize=40)

plt.legend(['High', 'Low'], fontsize=35)
plt.show()
```
